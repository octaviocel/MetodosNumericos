```julia
using Plots

deltaT=1e-2
u_=1.0
v_=0.0
T=3.15*2
n=630

```

```julia
function clase(dt,u0,v0,N)
    xn= Vector()
    un= fill(u0,N+1)
    vn= fill(v0,N+1)
    i=0
    while i<N
        un[i+1]=(un[i]-(dt*vn[i]))/(1+(dt^2))
        vn[i+1]= ((dt*un[i])+vn[i])/(1+(dt^2))
        append!(xn, ((i)*10^(-2)))
        i+=1
    end
    append!(xn,((N-1)*10^(-2)))
    append!(xn,((N)*10^(-2)))
    return un,vn,xn
end
```

```julia
u,v,x= clase(deltaT,u_,v_,n)
```

```julia
println(length(u))
println(length(v))
println(length(x))

primer = plot(x,v)
plot!(primer,x,u)
```
